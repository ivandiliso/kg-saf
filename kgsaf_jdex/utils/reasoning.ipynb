# Entailment and Materialization

```python
properties = [
    "SubClass",
    "EquivalentClass",
    #"DisjointClasses",
    #"DataPropertyCharacteristic",
    #"EquivalentDataProperties",
    #"SubDataProperty",
    #"ClassAssertion",
    #"PropertyAssertion",
    "EquivalentObjectProperty",
    "InverseObjectProperties",
    "ObjectPropertyCharacteristic",
    "SubObjectProperty",
    "ObjectPropertyRange",
    "ObjectPropertyDomain"
]

print("Using Axioms Generator on:")
prop_string = ""
for p in properties:
    print(f"\t{p}")
    prop_string += " " + p
```

```python
!java -Xmx20G -jar {robot_jar} reason -vvv \
  --reasoner HermiT \
  --input {onto_filename} \
  --output {onto_filename + "_reasoned.owl"} \
  --axiom-generators "{prop_string}" \
  --remove-redundant-subclass-axioms false \
  --exclude-tautologies structural \
  --include-indirect true \
  -D {onto_filename + "_debug.owl"}
```
