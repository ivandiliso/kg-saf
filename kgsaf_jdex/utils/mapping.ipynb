# MAPPING Entities to Unique IDS


```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

import argparse
from pathlib import Path
import utils.conventions.paths as pc
from utils.conventions.builtins import BUILTIN_URIS
from rdflib import Graph, RDF, RDFS, OWL, BNode
import json
import utils.conventions.paths as ptcn

base_path = Path("/home/navis/dev/kg-saf/kgsaf-data/datasets/materialize/YAGO4-20-C-MATERIALIZE")
```

```python
onto = Graph()
onto.parse(base_path / ptcn.ONTOLOGY)

print("Loaded Ontology")

ind_onto = Graph()
ind_onto.parse(base_path / ptcn.INDIVIDUALS)

print("Loaded Individuals")
```

```python
classes =  set(onto.subjects(RDF.type, OWL.Class)) - BUILTIN_URIS
classes = {c for c in classes if not isinstance(c, BNode)}

properties = set(onto.subjects(RDF.type, OWL.ObjectProperty)) - BUILTIN_URIS
individuals = set(ind_onto.subjects(RDF.type, OWL.NamedIndividual)) - BUILTIN_URIS

classes = list(classes)
properties = list(properties)
individuals = list(individuals)

classes.sort()
properties.sort()
individuals.sort()

print("Classes", len(classes))
print("Properties", len(properties))
print("Individuals", len(individuals))
```

```python
classes_mappings = {str(c):i for i,c in enumerate(classes)}
individuals_mappings = {str(c):i for i,c in enumerate(individuals)}
properties_mappings = {str(c):i for i,c in enumerate(properties)}

(base_path / ptcn.MAPPINGS).mkdir(exist_ok=True, parents=True)

with open(base_path / "mappings" / "class_to_id.json", "w") as f:
    json.dump(classes_mappings, f, indent=4)

with open(base_path / "mappings" / "individual_to_id.json", "w") as f:
    json.dump(individuals_mappings, f, indent=4)

with open(base_path / "mappings" / "object_property_to_id.json", "w") as f:
    json.dump(properties_mappings, f, indent=4)
```
